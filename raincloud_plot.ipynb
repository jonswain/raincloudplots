```python
import pandas as pd
from sklearn.datasets import load_diabetes, load_iris, load_wine

from raincloud_plot import raincloud_plot
```

```python
iris_dataset = load_iris()
iris_dataset = pd.DataFrame(iris_dataset["data"], columns=iris_dataset["feature_names"])

diabetes_dataset = load_diabetes()
diabetes_dataset = pd.DataFrame(
    diabetes_dataset["data"], columns=diabetes_dataset["feature_names"]
)

wine_dataset = load_wine()
wine_dataset = pd.DataFrame(wine_dataset["data"], columns=wine_dataset["feature_names"])
```

```python
raincloud_plot(
    iris_dataset,
    iris_dataset.columns,
    "Value",
    "Distribution of features in Iris Dataset",
    show_boxes=False,
)
```

```python
raincloud_plot(
    diabetes_dataset,
    diabetes_dataset.columns,
    "Value",
    "Distribution of features in Diabetes Dataset",
)
```

```python
raincloud_plot(
    wine_dataset,
    wine_dataset.columns,
    "Value",
    "Distribution of features in Wine Dataset",
    x_log_scale=True,
)
```
